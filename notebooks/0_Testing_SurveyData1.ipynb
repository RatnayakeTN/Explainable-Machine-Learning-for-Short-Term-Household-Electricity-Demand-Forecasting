```python
import numpy as np
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../data/raw/smart_6hour_1.csv")
```

```python
household_df = pd.read_csv("../data/processed/household_demographics.csv")
```

Get the time series records of surveyed households only

```python
df = df[
    df['household_ID'].isin(household_df['household_ID'])
]
```

```python
df.shape
```

```python
selected_ids = df['household_ID'].values
```

```python
files = [
    "../data/raw/smart_6hour_1.csv",
    "../data/raw/smart_6hour_2.csv",
    "../data/raw/smart_6hour_3.csv",
    "../data/raw/smart_6hour_4.csv",
    "../data/raw/smart_6hour_5.csv"
]

for i, file in enumerate(files):
    df = pd.read_csv(file)
    
    filtered_df = df[df['household_ID'].isin(selected_ids)]
    
    filtered_df.to_csv(f"../data/processed1/filtered_{i+1}.csv", index=False)
```

```python
f_files = [
    "../data/processed1/filtered_1.csv",
    "../data/processed1/filtered_2.csv",
    "../data/processed1/filtered_3.csv",
    "../data/processed1/filtered_4.csv",
    "../data/processed1/filtered_5.csv"
]
```

```python
df_list = [pd.read_csv(file) for file in f_files]
```

```python
final_df = pd.concat(df_list, ignore_index=True)
```
