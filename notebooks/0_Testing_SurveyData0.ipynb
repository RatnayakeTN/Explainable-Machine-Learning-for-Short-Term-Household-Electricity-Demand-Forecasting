```python
cols = [
    'household_ID',
    'member_ID',
    'age',
    'gender',
    'main_activity_engaged_in',
    'no_of_hours_stayed_at_home_during_last_week'
]
```

```python
w1 = pd.read_csv('../data/raw/w1_demographics.csv')[cols]

w2_new = pd.read_csv('../data/raw/w2_new_household_members.csv')[cols]

w3_new = pd.read_csv('../data/raw/w3_new_household_members.csv')[cols]
```

```python
survey_dates = pd.read_csv('../data/raw/survey_dates.csv')
```

```python
survey_dates['wave1'] = pd.to_datetime(survey_dates['wave1'], errors='coerce')
survey_dates['wave2'] = pd.to_datetime(survey_dates['wave2'], errors='coerce')
survey_dates['wave3'] = pd.to_datetime(survey_dates['wave3'], errors='coerce')
```

```python
w1['wave'] = 1
w2_new['wave'] = 2
w3_new['wave'] = 3
```

```python
wave1_ids = survey_dates[~survey_dates['wave1'].isna()]['household_ID']
wave2_ids = survey_dates[~survey_dates['wave2'].isna()]['household_ID']
wave3_ids = survey_dates[~survey_dates['wave3'].isna()]['household_ID']
```

```python
wave1_full = w1.copy()

wave1_full = wave1_full.merge(
    survey_dates[['household_ID', 'wave1']],
    on='household_ID',
    how='left'
)

wave1_full['survey_date'] = wave1_full['wave1']
wave1_full['wave'] = 1
wave1_full = wave1_full.drop(columns=['wave1'])
```

```python
wave2_full = pd.concat([w1, w2_new], ignore_index=True)

wave2_full = wave2_full.merge(
    survey_dates[['household_ID', 'wave2']],
    on='household_ID',
    how='left'
)

wave2_full['survey_date'] = wave2_full['wave2']
wave2_full['wave'] = 2

wave2_full = wave2_full.drop(columns=['wave2'])
```

```python
wave3_full = pd.concat([w1, w2_new, w3_new], ignore_index=True)

wave3_full = wave3_full.merge(
    survey_dates[['household_ID', 'wave3']],
    on='household_ID',
    how='left'
)

wave3_full['survey_date'] = wave3_full['wave3']
wave3_full['wave'] = 3

wave3_full = wave3_full.drop(columns=['wave3'])
```

```python
all_df = pd.concat([
    wave1_full,
    wave2_full,
    wave3_full
], ignore_index=True)
```

```python
all_df = all_df.drop_duplicates(
    subset=['household_ID', 'survey_date', 'member_ID']
)
```

```python
all_df['main_activity_engaged_in'].value_counts().to_frame()
```

```python
activity_map = {
    'Engaged in economic activity/ currently employed/ engaged in own business': 'employed',    
    'Household activities': 'not_employed',
    'Student': 'not_employed',
    'Too old / Disable/ unable to work': 'not_employed',
    'Retired and obtaining government/semi-government pension payment and is currently engaged in economic activity (employed elsewhere other than the place where he/she is receiving the pension from / engaged in own business)': 'employed',
    '1.Engaged in economic activity/ currently employed/ engaged in own business': 'employed',
    'Retired - Obtaining government/semi-government pension payment and currently not engaged in economic activity (not employed elsewhere or not engaged in any own business)': 'not_employed',
    'Retired from the private/semi-government sector and does not obtain any pension payment': 'not_employed',
    '7.Household activities': 'not_employed',
    'Seeking for and available to work': 'not_employed',
    'Received other pension payments': 'not_employed',
    '8.Student': 'not_employed',
    'Other': 'not_employed',
    '9.Too old / Disable/ unable to work.': 'not_employed',
    '2.Retired and obtaining government/semi-government pension payment and is currently engaged in economic activity (employ': 'employed',
    '3.Seeking for and available to work': 'not_employed',
    '6.Received other pension payments.': 'not_employed',
    '4.Retired - Obtaining government/semi-government pension payment and currently not engaged in economic activity (not emp': 'not_employed',
    '5.Retired from the private/semi-government sector and does not obtain any pension payment.': 'not_employed',
    '10.Other': 'not_employed'
}
```

```python
all_df['activity'] = all_df['main_activity_engaged_in'].map(activity_map)
```

```python
all_df.shape
```

```python
all_df['is_employed'] = (all_df['activity'] == 'employed').astype(int)
```

```python
all_df.head(15)
```

```python
all_df.isna().sum()
```

```python
all_df['is_child'] = (all_df['age'] < 18).astype(int)
all_df['is_elderly'] = (all_df['age'] >= 60).astype(int)
```

```python
household_df = all_df.groupby(['household_ID', 'survey_date']).agg({
    'member_ID': 'count',
    'age': ['mean', 'max'],
    'is_child': 'sum',
    'is_elderly': 'sum',
    'is_employed': 'sum',
    'no_of_hours_stayed_at_home_during_last_week': 'mean',
}).reset_index()
```

```python
household_df.columns = [
    'household_ID',
    'survey_date',
    'household_size',
    'avg_age',
    'max_age',
    'num_children',
    'num_elderly',
    'num_employed',
    'avg_hours_home',
]
```

```python
household_df['dependency_ratio'] = (
    (household_df['num_children'] + household_df['num_elderly']) /
    household_df['household_size']
)
```

```python
household_df['employment_ratio'] = (
    household_df['num_employed'] / household_df['household_size']
)
```

```python
household_df['num_adults'] = (
    household_df['household_size'] -
    household_df['num_children']
)
```
