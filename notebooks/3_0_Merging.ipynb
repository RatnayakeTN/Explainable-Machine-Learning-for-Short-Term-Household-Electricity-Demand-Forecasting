```python
household_df_reorder = household_df_reorder.rename(columns={'survey_date': 'datetime'})
```

```python
processed_selection['datetime'] = pd.to_datetime(processed_selection['datetime'])
appliances['datetime'] = pd.to_datetime(appliances['datetime'])
household_df_reorder['datetime'] = pd.to_datetime(household_df_reorder['datetime'])
```

```python
processed_selection = processed_selection.sort_values(['household_ID', 'datetime'])
appliances = appliances.sort_values(['household_ID', 'datetime'])
household_df_reorder = household_df_reorder.sort_values(['household_ID', 'datetime'])
```

Merge appliances and demographics

```python
survey_features = appliances.merge(
    household_df_reorder,
    on=['household_ID', 'datetime'],
    how='left'
)
```

Time aware merge into 6 hour time series

```python
processed_selection['datetime'] = pd.to_datetime(processed_selection['datetime'], errors='coerce')
survey_features['datetime'] = pd.to_datetime(survey_features['datetime'], errors='coerce')
```

```python
# Drop NaT
processed_selection = processed_selection.dropna(subset=['datetime'])
survey_features = survey_features.dropna(subset=['datetime'])
```

```python
processed_selection = processed_selection.sort_values(['household_ID', 'datetime'], kind='mergesort')
survey_features = survey_features.sort_values(['household_ID', 'datetime'], kind='mergesort')
```

```python
processed_selection = processed_selection.sort_values('datetime')
survey_features = survey_features.sort_values('datetime')
```

```python
processed_selection = processed_selection.reset_index(drop=True)
survey_features = survey_features.reset_index(drop=True)
```

```python
full_features = pd.merge_asof(
    processed_selection,
    survey_features,
    by='household_ID',
    left_on='datetime',
    right_on='datetime',
    direction='backward'
)
```

```python
full_features.to_csv("../data/processed1/final_selection.csv", index=False)
```

```python
full_survey = full_features.copy()
```

```python
full_survey = full_survey.sort_values(['household_ID','datetime'])
```

```python
full_survey.head()
```

```python
full_survey.shape
```

```python
survey_cols = [											
    'total_appliances',
    'high_energy_appliances',
    'Heating_Cooking',
    'Cooling_Refrigeration',
    'Security_Infrastructure',
    'Cleaning_Household',
    'Entertainment_ICT',
    'Communication_IT',
    'Mobility_Energy',
    'Other',
    'cooling_ratio',
    'digital_lifestyle',
    'appliance_diversity',
    'household_size',
    'avg_age',
    'max_age',
    'num_children',
    'num_adults',
    'num_elderly',
    'num_employed',
    'avg_hours_home',
    'dependency_ratio',
    'employment_ratio'
]

full_survey[survey_cols] = (
    full_survey
    .groupby('household_ID')[survey_cols]
    .ffill()
)
```

```python
full_survey['valid_start'] = (
    full_survey
    .groupby('household_ID')['total_appliances']
    .transform(lambda x: x.notna().cumsum() > 0)
)
```

```python
full_survey = full_survey[full_survey['valid_start']].copy()
```

```python
full_survey.drop(columns=['valid_start'], inplace=True)
```

```python
full_survey.groupby('household_ID')['datetime'].diff().dropna().value_counts().head()
```

```python
full_survey.groupby('household_ID').agg(
    start=('datetime', 'min'),
    end=('datetime', 'max'),
    count=('datetime', 'size')
).describe()
```

#### Valid Households

Handle households which may start very late which could be useless when modelling

```python
min_points = 100 

valid_households = (
    full_survey
    .groupby('household_ID')
    .filter(lambda x: len(x) >= min_points)
)
```
